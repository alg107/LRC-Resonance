## Introduction

Oscillatory motion arises when there is a force on an object that grows in proportion to its displacement from some point. The most commonly given example of this phenomena is a mass on a spring but this motion can be observed in many different contexts. More specifically, coupled oscillators can be used to model the behaviour of many real world systems. Most notably, the behaviour of solids which are made up of atoms coupled together by strong electrostatic forces. Therefore, the insights we can gain from a simple setup of masses and springs can have far-reaching consequences for our understanding of more complex topics. One key insight gained from the mathematics of oscillatory systems is the idea of normal modes. If one was to observe the motion of each of the masses of a coupled oscillator system individually, the paths they take would look chaotic and almost random, but by resolving these motions into normal modes. One can look at the system as a whole and explain this chaotic motion.

To represent the idea of coupled oscillators, say we have N equally-spaced, coupled oscillators of mass $M$ oscillating longitudinally between two fixed points. The spring constant of each of the coupling springs is identical and denoted the force constant $K$. The equilibrium positions of each mass are therefore given by $x_n = na$ where $n = 1,2,...,N$ and $a$ is the distance between each of the masses. If we denote the displacement of each mass from its equilibrium $u_n$, the equation of motion of the $n^{th}$ mass, considering the masses on either side, is given by 

$Mü_n = -K(u_n-u_{n-1}) - K(u_n-u_{n+1})$.

Simplifying, 

$Mü = Ku_{n-1} - 2Ku_n + Ku_{n+1}$.

We take that $u_0$ and $u_{n+1}$ are equal to 0 as they represent the fixed points of the system. Therefore, we look for a standing-wave solution according to the trial solution of form 

$u_n = Acos(\omega t)sin(\frac{2 \pi x_n}{\lambda}) = Acos(\omega t)sin(\frac{2 \pi n a}{\lambda})$.

If we substitute this trial solution into the equation of motion, we will find a relationship between $\omega$ and and $\lambda$. As we have the fixed points at $u_0$ and $u_{n+1}$, we require that $2\pi(N+1)a/\lambda = m\pi$ where m is an integer. This ensures the standing wave will 'fit' with the conditions we have imposed. Substituting we find

$\omega = 2\sqrt{\frac{K}{M}}sin(\frac{\pi a}{\lambda})$

Therefore, for each value of $\lambda_m$ there is an associated value of $\omega_m$ given by 

$\omega_m = 2\sqrt{\frac{K}{M}}sin(\frac{\pi a}{\lambda_m}) = 2\sqrt{\frac{K}{M}}sin(\frac{m \pi}{2(N+1)})$

If we substitute in a value of $m = N + 1$, we find that we get a value of $\lambda = 2a$ and therefore $u_n = 0$ for all masses. Therefore we should consider only values of $m$ such that $m = 1,2,...,N$. 

This relationship between wavelength and angular frequency is called a dispersion relation and is important in calculating phase and group velocity of modes of a given oscillatory system. 

The objective of this experiment was to investigate normal-mode frequencies of a system of coupled oscillators. From the amplitudes of disaplacement of these modes a dispersion relation was calculated and used to find values for phase and group velocity for each of the 9 modes.

## Method

### Calculation of Theoretical Mode Frequencies

Before any practical measurements were carried out, the normal mode frequencies of the system of masses had to be determined. This was done using the following equation derived above.

$\omega_m = 2\sqrt{\frac{K}{M}}sin(\frac{m \pi}{2(N+1)})$

This gives us an angular frequency for each mode $m$ of the system where $m = 1,2,...,9$ and N is the number of masses which is $9$. In actual fact, the experimental system consisted of 10 coupled oscillators with one mass being directly driven by a driving force of angular frequency $\omega$, however we considered only the 9 masses not directly driven between the fixed point on the non-driven end and the node between the 9th and 10th masses, allowing us to use this theoretical model. We then converted these angular frequencies to frequencies using the relation $\lambda = \omega / 2\pi$. This left us with a list of 9 mode frequencies.

### Calculating Average Spring Constant

Using a retort stand, we hung up a spring and attached a weight set with a removable piece to it, using a ruler to measure the height of the weight above the table. We then detached the removable piece from the weight and measured its new height above the table. Using the difference between the two heights as our $\Delta x$, the mass of the detached piece as our $m$, we calculated the spring constant by equating the forces on the mass.

$F = -Kx = Mg$

Therefore, the spring constant is given by 

$K = \frac{mg}{\Delta x}$.

We repeated the process for all ten springs individually, and used the mean of the ten constants in further calculations.

### Measurement of Single Car Oscillation Frequency

The experimental setup consisted of 10 masses of recorded weight coupled by springs of equal spring constant on an air track through which an industrial vacuum cleaner was blowing air at 120 $V$, resulting in a near frictionless surface. The right-most mass was driven by a driving motor of variable frequency which could be adjusted to within 0.01 $Hz$. Before attempting to find any of the modes of this system, we first verified that a single car oscillates as expected. Each car is connected to 2 springs of equal spring constant and if the other ends of these springs remained fixed, the cars angular frequency is given by $\omega = \sqrt{2K/M}$. To confirm this, the 8th oscillator was held in place and as the 10th oscillator is held in place by its driving motor, the 9th oscillator was now between two fixed points. It was displaced roughly 5cm from its equilibrium position and timing began once it had reached the position at which it was displaced after one full oscillation. Timing was then stopped after 10 oscillations had taken place and the mass reached the same position at which it was displaced. This time was recorded as well as the number of oscillations that had taken place. From this we could proceed to confirm that each of the masses oscillates as expected.

### Finding the Wavelength of Each Mode

As we had collected the mode frequencies of the system and confirmed that each car oscillates as expected, we could proceed to find and classify the mode frequencies. Beginning with the lowest frequency in our list of calculated mode frequencies, we set the driving motor frequency to this frequency. We expected that this mode would have a wavelength of twice the length of the air track and so we used this knowledge to make small adjustments to the frequency until the standing wave became clear. Once this first mode was found, the exact frequency at which it was found was recorded and the positive and negative displacement for each of the masses was recorded, taking into account the phase of each mass with respect to the other masses. The phase data was recorded implicitly in the order of the positive and negative displacements. The measurement of the displacements was carried out by 3 people as fast as possible to reduce the effects of a change in amplitude due to the effect of beating from the driving frequency being very close but not exactly on the mode frequency. These steps were then repeated for the remaining 8 mode frequencies. We also took into account the effect of a sudden change in forcing frequency in creating a transient motion. By waiting 30 seconds after the final change in frequency before measuring displacements, we could reduce the effect of this negative exponential factor on our measured amplitudes. 


## Analysis

### Setup and helper function definition

```python
# Setup and helper function definition

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pds
from scipy.optimize import minimize
from scipy.stats import chisquare
from decimal import Decimal

# Rounds uncertainties to one significant figure
def round_to_1(x):
    return round(x, -int(np.floor(np.log10(abs(x)))))

# Calculates amplitudes from initial and final displacements
def pairs_amps(lst):
    if (len(lst)%2)!=0:
        raise Exception("Odd list")
    amps = []
    for i, val in enumerate(lst):
        if (i%2)==0:
            amps.append((lst[i+1]-lst[i])/2)
    return amps


# Model for a sinusoid
def sinusoid(params, t):
    A, w, phase = params
    return A*np.sin(w*t+phase)


# Function to compute chi-squared
def chi2(params, data):

    t = data[:,0]
    y = data[:,1]
    #sigma = data [:,2]
    
    expected = sinusoid(params, t)
    return chisquare(y, expected)[0]

    # Add some lines to compute and return chi^2.
    # Make use of your mlens function.

# Fits data points to sinusoidal model
def calc_sin(guesses, pnts):
    result = minimize(chi2,guesses, method='BFGS',args=(pnts))
    param_unc = np.sqrt(2*np.diag(result.hess_inv))
    newparams = result.x
    #print(newparams)
    return newparams, param_unc[1]


# Unit conversion lambdas
in_to_cm = lambda x: 2.54*x
w_to_f = lambda w: w/(2*np.pi)

# Importing the data
data = pds.read_csv("air_track.csv")
```

### Measuring the Spring Constant

```python
extensions = (np.array(data.initial - data.final)*1e-2)[:10]

# Constant definition
m = 19.99e-3 # 19.99 grams (Testing mass)
m_car = 0.184 # 0.184 kg (Car mass)
cardist = 10e-2 # 10 cm (Distance between each car)
g = 9.81

# calculating extensions and uncertainty
del_x = np.mean(extensions)
del_x_unc = np.std(extensions)/np.sqrt(len(extensions))

# Calculating the spring constant
s = (m*g)/del_x
# Propagation of uncertainties
s_unc = s*(del_x_unc/del_x)

print("Calculated Spring Constant: ", round(Decimal(s),2),"±", round_to_1(s_unc),  "Nm^-1")
```

### Calculating the Theoretical Mode Frequencies

```python
w = lambda m, K, M: 2*np.sqrt(K/M)*np.sin((np.pi*m)/20)

fqs = []
print("Theoretical Mode Frequencies:")
for i in range(9):
    f = Decimal(w_to_f(w(i+1, s, m_car)))
    fqs.append(f)
    print("f", i+1, ":", round(f,4), "Hz")
#plt.plot(fqs, marker="s", linestyle="none")
```

### Measurement of Single Car Oscillation Frequency

```python
expected_w = np.sqrt((2*s)/(m_car))
print("Expected frequency for one car: ", round(w_to_f(expected_w),4), "Hz")

measured_period_10osc = 9.0 # Measured time in seconds for 10 oscillations of a single car
reaction_time = 0.2*2
measured_period = measured_period_10osc/10

observed_f = 1/measured_period

# Propagation of uncertainties
f_unc = observed_f*(reaction_time/measured_period_10osc)
print("Observed frequency for one car: ", round(Decimal(observed_f), 4), "±", round_to_1(f_unc), "Hz")
```

### Finding the Wavelength of Each Mode

```python
theo_lambdas = [(2*(9+1)*cardist)/m for m in range(1,10)]
sinusoid_amplitudes = np.array([16,15,5,5,1,1,0.6,0.7,0.5])/100
sinusoid_amplitudes2 = np.array([16,15,5,5,1,1,1,0.5,1])/100

exp_lambdas = []
exp_lambdas_unc = []

amplitude_measurement_uncertainty = 0.5e-2 # (1cm)
frequency_measurement_uncertainty = 0.01 # According to measurement equipment

for i in range(0,9): 
    plt.figure(i+1)
    plt.title("Mode "+str(i+1))
    plt.xlabel("Distance (m)")
    plt.ylabel("Displacement (m)")
    amps = np.array(pairs_amps(data["M"+str(i+1)]))/100
    amps_uncs = (amps**0)*amplitude_measurement_uncertainty
    xs = cardist*np.array(range(1,10))
    #xs = np.array(theo_lambdas)
    xs_uncs = (xs**0)*frequency_measurement_uncertainty
    plt.errorbar(xs,amps, amps_uncs, xs_uncs, linestyle="None", marker="s", label="Observed Amplitudes")
    #a, a_unc = bestfit(xs, amps,amps_uncs, xs_uncs)
    xx = np.linspace(0,cardist*10,100)
    
    # Plots sine curve based on chi square minimisation
    guesses = (sinusoid_amplitudes[i], (2*np.pi)/theo_lambdas[i], 0)
    #print("Guesses: ", guesses)

    params, calc_w_unc = calc_sin(guesses, np.column_stack((xs, amps)))
    #print("Params: ", params)
    exp_lambdas.append(params[1])
    exp_lambdas_unc.append(calc_w_unc)
    plt.plot(xx, sinusoid((guesses[0]*(0.9+((i+1)%2)*0.2), params[1], params[2]),xx), label="Fitted Sinusoidal Model")
    
    ## Plots sine curve according to theoretical standing wave
    plt.plot(xx, sinusoid((sinusoid_amplitudes[i],(2*np.pi)/theo_lambdas[i], 0), xx), linestyle=":", label="Theoretical Standing Wave")
    plt.legend()
    

w_to_lambda = lambda w: (2*np.pi)/w
```

## Discussion

The main source of uncertainty in this experiment was introduced in the measurement of the displacements of each mass for each mode frequency. This experiment relied on the assumption that the amplitudes of each mass would remain constant as a function of time. However, this is not true. The ratios of the amplitudes of each mass to the other masses remains constant, but the absolute values of these amplitudes tended to increase and decrease over time in an oscillatory fashion. This led us to believe that this was due to a beating phenomena. When two waves of similar but not equal wavelength are superimposed on each other, there is a resulting oscillation of the amplitude from the sum of the individual amplitudes to 0. The frequency of this oscillation is determined by the absolute difference between the frequencies of each individual wave. If it were true that beating phenomena was responsible for the oscillation of amplitude in this experiment, there would be two ways to reduce this effect. The first is a strategy we attempted to employ which is taking the measurements as fast as possible. If all displacement measurements were somehow taken instantaneously, the effect of the beating phenomena would be completely minimised, but as this is not possible or at least not practicable in this context, we could reduce this effect by taking measurements as fast as possible. However by having 3 people carry out the measurements, uncertainty can be introduced through a lack of communication as each person may employ a different strategy in measuring displacement. Another thing to consider is that the phase of each mass was not recorded explicitly, but rather implicitly in the order of the initial and final displacements. With 3 people recording these displacements this could lead to a confusion of phases, resulting in invalid data. The second way to reduce the effect of beating is to reduce the difference in frequency between the driving frequency and real mode frequency. Therefore, if we were to repeat this experiment, we would likely be more precise in finding the mode frequencies of the system and to measure the displacements, we would setup a camera to take two picture $\pi$ radians apart to measure the initial and final displacements with minimal change in absolute amplitude of the system. 

In using the measurements of displacements for each mode to calculate the amplitudes of each mass's oscillation and then plot the amplitude with respect to its distance along the air track, it was found that the sinusoidal fit applied to the data points did not exactly match the theoretical standing wave that should be present. For modes 1,2,4,5,6, and 8, these discrepancies were minimal. However, for modes 3,7, and 9, these discrepancies were larger. It should be noted that there are three ways in which the sinusoidal fit can differ from the theoretical standing wave corresponding to the 3 variables that make up a sinusoidal curve. The first is amplitude. However, the amplitude is not particularly important and will vary depending on the absolute amplitude of each mass. The second variable is phase shift. Phase shift is also not particularly important in determining dispersion relation however it can be a good indicator of whether or not a fitted sinusoidal function is a good match to the data and theoretical standing wave. We see that in the case of mode 3, the phase shift is significantly different to the theoretical standing wave but the wavelength is very similar and so this model is still a good fit for our purposes in calculating a dispersion relation. However, it is clear that with modes 7 and 9, the wavelengths do not line up with the theoretical standing wave and thus these data points will have an affect on the experimentally calculated dispersion relation. It is not clear what the source of these discrepancies is. One possibility is the measurement of displacements. As mode frequency increases, amplitudes decrease and thus it becomes harder to measure displacement resulting in larger uncertainties that could account for a poor sinusoidal representation. Another possible explanation is the sinusoidal fitting function itself. A sine function with variable frequency is not a linear function and so the usual methods of fitting functions of linear regression were not able to be used. For non-linear models a method of minimisation of using the chi-square can be used and was used in this experiment. This method requires an initial guess at the optimised parameters and different guesses can lead to different parameters being produced by the algorithm. Therefore, the initial guesses used in this experiment could have introduced a bias into the sinusoidal modelling function, leading to the discrepancies in modes 3, 7, and 9. Therefore, if we were to repeat this experiment, we would likely analyse different initial guesses for this function and observe the effects of these guesses on the computed parameters to investigate how changes to these guesses affect our final dispersion relation. 

A major overlooked factor in this experiment was the effect of damping on the amplitude and oscillating frequency. Although damping does not have a large effect on the oscillation frequency for such lightly damped systems, it nonetheless has an effect which can be calculated using the relationship $\omega_d = \sqrt{\omega_0^2 - \mu^2}$. This frequency shift was not taken into account in this experiment and although it is unlikely that this would have an observable effect on our final results, at the very least, the effect of damping on the oscillation frequency should have been investigated to confirm this fact. Damping also has an effect on the amplitude of oscillation according the relation

$A = A_0e^{-\mu t}$

where $\mu$ is the damping factor. Damping removes energy from the system and therefore if this were the only change in energy of the system, we would not get the ongoing oscillation that is observed. We also have the driving motor which puts energy into the system. If the system were undamped, at the normal modes, the amplitude would continue to grow as more energy is put into the system. However as both of these energy processes occur in this system. Assuming the driving power is not that much greater than the power changed through damping, the system should come to an energy equilibrium in which the amplitude reaches a steady value other than the effect of beating. As damping is given by a negative exponential relation with time we would expect that this equilibrium would essentially occur after a given time and therefore this emphasises the importance of waiting at least 30 seconds for any transient factors in the system to minimise their effect.

Not looking at each part of a system individually but rather considering the larger patterns as they affect the sum of these parts.
